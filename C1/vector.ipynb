```python
import time
```

```python
a = np.random.rand(10000000)
b = np.random.rand(10000000)
tic = time.time()
c = np.dot(a,b)
toc = time.time()
print(c)
print("Vectorized Time "+ str(1000*(toc-tic))+"ms")
```

```python
c=0
tic = time.time()
for i in range(10000000):
  c+=a[i]*b[i]

toc = time.time()
print(c)
print("For Loop  Time "+ str(1000*(toc-tic))+"ms")
```
