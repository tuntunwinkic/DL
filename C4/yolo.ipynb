```python
import tensorflow as tf
```

```python
tf.random.set_seed(10)
box_confidence = tf.random.normal([19, 19, 5, 1], mean=1, stddev=4, seed = 1) # box ငါးခု အတွက် pc တခု စီ
boxes = tf.random.normal([19, 19, 5, 4], mean=1, stddev=4, seed = 1) # box 5 ခုမှာ bx,by,bh,bw ရှိ
box_class_probs = tf.random.normal([19, 19, 5, 80], mean=1, stddev=4, seed = 1) #box ငါးခုမှာ တခုကို class 80 ရှိ
```

```python
box_scores = box_confidence * box_class_probs # class 80 ကို confidence နဲ့ သွားမြှောက်

```

```python

box_classes = tf.math.argmax(box_scores, axis=-1)#[19, 19, 5, 80] ထဲက အခု ၈၀ ထဲမှာ ဘယ် class က prob အမြင့်ဆုံးလဲ အဲ့ အမြင့်ဆုံး class ကိုပဲ​ယူ

# ဘယ် class လဲပေါ့ class အခု ၈၀ ထဲက ဘာ class လဲ ပြောတာ

box_class_scores = tf.math.reduce_max(box_scores,axis=-1)# 80 ထဲမှာ prob အမြင့်ဆုံး score ကို ယူမယ်
# အဲ့ အမြင့်ဆုံး score က ဘယ်လောက်လဲပေါ့
```

```python
box_classes.shape
```

```python
box_class_scores.shape
```

```python
threshold = 0.5
```

```python
filtering_mask = box_class_scores >= threshold
```

```python
filtering_mask.shape
```

```python
scores = tf.boolean_mask(box_class_scores, filtering_mask) # threshold ထက်ကြီးတဲ့ score တွေပဲ​ယူ
boxes = tf.boolean_mask(boxes, filtering_mask) # threshold ထက်ကြီးတဲ့ box  တွေပဲ​ယူ
classes = tf.boolean_mask(box_classes, filtering_mask) # threshold ထက်ကြီးတဲ့ classes တွေပဲ​ယူ
```
